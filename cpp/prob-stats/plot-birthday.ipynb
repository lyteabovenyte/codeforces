```python
import math
import matplotlib.pyplot as plt
```

```python
# Birthday paradox: plot probability that at least two people share a birthday vs. group size.
# Assumptions: 365 equally likely birthdays, no leap day.

DAYS = 365
N_MAX = 100  # plot up to 100 people

p_no = 1.0  # probability all birthdays are distinct
ns = []
p_shared_list = []
min_n = None

for n in range(1, N_MAX + 1):
    p_no *= (DAYS - (n - 1)) / DAYS
    p_shared = 1.0 - p_no
    ns.append(n)
    p_shared_list.append(p_shared)
    if min_n is None and p_shared >= 0.5:
        min_n = n
        min_p = p_shared

print(f"Minimum n for >= 50% chance: n = {min_n}, P = {min_p:.6f}")

plt.figure(figsize=(8,5))
plt.plot(ns, p_shared_list, linewidth=2)
plt.axhline(0.5, linestyle="--")
if min_n is not None:
    plt.axvline(min_n, linestyle="--")
    plt.annotate(
        f"n = {min_n}\nP ≈ {min_p:.3f}",
        xy=(min_n, min_p),
        xytext=(min_n + 5, min_p - 0.1),
        arrowprops=dict(arrowstyle="->")
    )
plt.xlabel("Number of people in the room (n)")
plt.ylabel("P(at least one shared birthday)")
plt.title("Birthday Paradox (365 Days, No Leap Day)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
